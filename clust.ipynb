```python
import csv
import numpy as np
from fuzzywuzzy import fuzz
f=open('data.csv')
r=csv.reader(f)
data=[]
dob=[]
train=[]
test=[]
a=[]
output=[] # 2D array to collect unique patient detail 
clust=[]   # araay to store the cluster 
j=0




#function to calculate score of matching data 
def comp(a,b):
    count=0
    if a[1]==b[1]:
        count=count+1
    count=count+2*fuzz.ratio(a[0]+a[3],b[0]+b[3])/100
    return count
for row in r:
    data.append(row)
    

# testing and training data separation     
for k in data:
    if j%7==0:
        test.append(k)
    else:
        train.append(k)
    j=j+1


#cluster initialization with one person's detail 
a.append(train[0])
clust.append(a)
a=[]
del train[0]   # deleting that data from the training so that model


# main model to train and clustering 
# having less data this code is using same model for testing also 
# 100 is the limit of sample data for the one cluster 
def model(train,clust):
    for k in train:
        a=[]
        done=0
        for j in clust:
            count=0
            for m in j:
                count=count+comp(k,m)
            count=count/len(j)
            if count>2.4 and len(j)<100:
                j.append(k)
                done=1
        if done ==0:
            a.append(k)
            done=1
            clust.append(a)
            a=[]
    return clust


clust=model(train,clust) # calling the model on train data 
clust=model(test,clust)  # calling the model on testing data 

#keeping  one person's detail from every cluster as a unique patient
output.append(['fn','dob','gn','ln'])
for j in clust:
    output.append(j[0])

    
print(len(output))   # printing nos of unique patients

#printing  output array
print(*output,sep='\n\n')


```

```python
#creating csv file with output 
with open(r'output1.csv','a') as f:
    writer=csv.writer(f)
    for k in output:
        writer.writerow(k)
        
```
